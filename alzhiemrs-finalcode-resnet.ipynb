```python
"""
COMPLETE END-TO-END ALZHEIMER'S MRI CLASSIFICATION WITH RESNET18
Using Hugging Face Dataset: Falah/Alzheimer_MRI
Includes: Data Loading -> Preprocessing -> Training -> Hyperparameter Tuning -> Evaluation

NOTE: Keeps your original preprocessing, learning rates, and optimizers.
Only adds combinations over dropout and weight decay (NO scheduler).
Also adds F2 score in the final evaluation.
"""

import numpy as np
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torchvision import models
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, fbeta_score
import copy
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# Hugging Face datasets
from datasets import load_dataset

print("="*70)
print("ALZHEIMER'S MRI CLASSIFICATION - COMPLETE END-TO-END PIPELINE")
print("Using Hugging Face Dataset: Falah/Alzheimer_MRI")
print("="*70)

# ==================== 1. DATA LOADING ====================
print("\n[STEP 1/6] Loading Data from Hugging Face...")

# Load the dataset from Hugging Face
print("Downloading dataset from Hugging Face Hub...")
dataset = load_dataset('Falah/Alzheimer_MRI')

print(f"\nDataset loaded successfully!")
print(f"Available splits: {list(dataset.keys())}")
print(f"Train samples: {len(dataset['train'])}")
print(f"Test samples: {len(dataset['test'])}")

# Define the mapping of the image labels
labelMap = {
    0: "Mild_Demented",
    1: "Moderate_Demented",
    2: "Non_Demented",
    3: "Very_Mild_Demented"
}

# Examine sample data
print("\nSample data structure:")
sample = dataset['train'][0]
print(f"Keys: {sample.keys()}")
print(f"Image type: {type(sample['image'])}")
print(f"Label: {sample['label']} -> {labelMap[sample['label']]}")

# ==================== 2. CREATE VALIDATION SPLIT ====================
print("\n[STEP 2/6] Creating Validation Split...")

# Split test set into validation and test (KEEPING YOUR ORIGINAL SPLIT LOGIC)
test_dataset = dataset['test']
total_test = len(test_dataset)
val_size = total_test // 2

# Create validation and test splits
val_dataset = test_dataset.select(range(val_size))
test_dataset = test_dataset.select(range(val_size, total_test))

print(f"Training samples: {len(dataset['train'])}")
print(f"Validation samples: {len(val_dataset)}")
print(f"Testing samples: {len(test_dataset)}")

# ==================== 3. CUSTOM DATASET CLASS ====================
print("\n[STEP 3/6] Creating Custom Dataset and Preprocessing...")

class AlzheimerDataset(Dataset):
    """Custom Dataset for Alzheimer MRI images"""
    def __init__(self, hf_dataset, transform=None):
        self.dataset = hf_dataset
        self.transform = transform
    
    def __len__(self):
        return len(self.dataset)
    
    def __getitem__(self, idx):
        sample = self.dataset[idx]
        image = sample['image']
        if image.mode != 'RGB':
            image = image.convert('RGB')
        if self.transform:
            image = self.transform(image)
        label = torch.tensor(sample['label'], dtype=torch.long)
        return image, label

# --------- KEEPING YOUR ORIGINAL PREPROCESSING ---------
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Create datasets
train_dataset_pytorch = AlzheimerDataset(dataset['train'], transform=transform)
val_dataset_pytorch   = AlzheimerDataset(val_dataset,   transform=transform)
test_dataset_pytorch  = AlzheimerDataset(test_dataset,  transform=transform)

print(f"PyTorch datasets created successfully!")

# ==================== 4. CREATE DATALOADERS ====================
print("\n[STEP 4/6] Creating DataLoaders...")

# Same batch size and workers as your first script
trainLoader   = DataLoader(train_dataset_pytorch, batch_size=32, shuffle=True,  num_workers=2)
valLoader     = DataLoader(val_dataset_pytorch,   batch_size=32, shuffle=False, num_workers=2)
testingLoader = DataLoader(test_dataset_pytorch,  batch_size=32, shuffle=False, num_workers=2)

print(f"Train batches: {len(trainLoader)}")
print(f"Validation batches: {len(valLoader)}")
print(f"Test batches: {len(testingLoader)}")

# ==================== VISUALIZE SAMPLE IMAGES ====================
print("\nVisualizing sample training images...")

# Get a batch
data_iter = iter(trainLoader)
images_sample, labels_sample = next(data_iter)

fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.ravel()
for i in range(8):
    img = images_sample[i].numpy().transpose(1, 2, 0)
    # Denormalize for display
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    img = std * img + mean
    img = np.clip(img, 0, 1)
    axes[i].imshow(img)
    axes[i].axis('off')
    axes[i].set_title(f"Label: {labelMap[labels_sample[i].item()]}", fontsize=10)
plt.suptitle("Sample Training Images", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Class distribution
print("\nClass Distribution:")
def get_class_distribution(hf_dataset):
    labels = [sample['label'] for sample in hf_dataset]
    distribution = {}
    for i, class_name in labelMap.items():
        count = labels.count(i)
        percentage = count / len(labels) * 100
        distribution[class_name] = (count, percentage)
    return distribution

for split_name, hf_dataset in [("Train", dataset['train']), ("Val", val_dataset), ("Test", test_dataset)]:
    print(f"\n{split_name}:")
    dist = get_class_distribution(hf_dataset)
    for class_name, (count, percentage) in dist.items():
        print(f"  {class_name}: {count} ({percentage:.1f}%)")

# Visualize class distribution
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for idx, (split_name, hf_dataset) in enumerate([("Train", dataset['train']),
                                                ("Val", val_dataset),
                                                ("Test", test_dataset)]):
    dist = get_class_distribution(hf_dataset)
    classes = list(dist.keys())
    counts = [dist[c][0] for c in classes]
    axes[idx].bar(classes, counts, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A'], alpha=0.8)
    axes[idx].set_title(f'{split_name} Set Distribution', fontsize=12, fontweight='bold')
    axes[idx].set_ylabel('Count', fontsize=10)
    axes[idx].tick_params(axis='x', rotation=45)
    axes[idx].grid(True, alpha=0.3, axis='y')
    for i, count in enumerate(counts):
        axes[idx].text(i, count + 10, str(count), ha='center', fontsize=9, fontweight='bold')
plt.tight_layout()
plt.show()

# ==================== 5. MODEL SETUP ====================
print("\n[STEP 5/6] Setting up ResNet18 Model and Training Pipeline...")

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

def create_resnet18_model(num_classes=4, pretrained=True, dropout=0.0):
    """Create ResNet18 model with optional dropout before the final layer."""
    model = models.resnet18(pretrained=pretrained)
    num_features = model.fc.in_features
    if dropout and dropout > 0:
        model.fc = nn.Sequential(nn.Dropout(p=dropout), nn.Linear(num_features, num_classes))
    else:
        model.fc = nn.Linear(num_features, num_classes)
    return model.to(device)

# ==================== TRAINING FUNCTIONS ====================
def train_epoch(model, dataloader, criterion, optimizer, device):
    """Train for one epoch"""
    model.train()
    running_loss, correct, total = 0.0, 0, 0
    for images, labels in tqdm(dataloader, desc="Training", leave=False):
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * images.size(0)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
    epoch_loss = running_loss / total
    epoch_acc = correct / total
    return epoch_loss, epoch_acc

def validate_epoch(model, dataloader, criterion, device):
    """Validate for one epoch"""
    model.eval()
    running_loss, correct, total = 0.0, 0, 0
    with torch.no_grad():
        for images, labels in tqdm(dataloader, desc="Validation", leave=False):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            running_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    epoch_loss = running_loss / total
    epoch_acc = correct / total
    return epoch_loss, epoch_acc

def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs, device,
                scheduler=None, scheduler_type='none'):
    """
    Complete training loop with history tracking.
    (scheduler kept in signature for compatibility; always None in this version)
    """
    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0
    
    for epoch in range(num_epochs):
        print(f'\nEpoch {epoch+1}/{num_epochs}')
        print('-' * 50)
        train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)
        val_loss, val_acc = validate_epoch(model, val_loader, criterion, device)
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        print(f'Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f}')
        print(f'Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}')
        # No scheduler stepping (scheduler is None)
        if val_acc > best_acc:
            best_acc = val_acc
            best_model_wts = copy.deepcopy(model.state_dict())
            print(f'✓ New best model! Val Acc: {best_acc:.4f}')
    model.load_state_dict(best_model_wts)
    return model, history

# ==================== HYPERPARAMETER EXPERIMENTATION (NO SCHEDULER) ====================
def hyperparameter_search(train_loader, val_loader, device, num_epochs=10):
    """
    Keep your learning_rates and optimizers,
    cross ONLY with dropout and weight_decay.
    (No scheduler; 24 runs total)
    """
    experiments = []
    
    # Your original choices:
    learning_rates = [0.001, 0.0001, 0.00001]
    optimizers_config = ['Adam', 'SGD']

    # New knobs:
    dropouts = [0.0, 0.3]
    weight_decays = [0.0, 1e-4]

    # 3 * 2 * 2 * 2 = 24 runs
    for lr in learning_rates:
        for opt_name in optimizers_config:
            for wd in weight_decays:
                for dp in dropouts:
                    experiments.append({
                        'lr': lr,
                        'optimizer': opt_name,
                        'weight_decay': wd,
                        'dropout': dp,
                        'scheduler': 'none',  # kept for printing consistency
                        'name': f'{opt_name}_lr{lr}_wd{wd}_dp{dp}_none'
                    })
    
    results = []
    all_histories = {}
    
    print(f"\n{'='*70}")
    print(f"HYPERPARAMETER SEARCH (no scheduler): {len(experiments)} EXPERIMENTS")
    print(f"{'='*70}\n")
    
    for idx, exp in enumerate(experiments):
        print(f"\n{'='*70}")
        print(f"Experiment {idx+1}/{len(experiments)}: {exp['name']}")
        print(f"LR={exp['lr']} | OPT={exp['optimizer']} | WD={exp['weight_decay']} | DP={exp['dropout']} | SCH=none")
        print(f"{'='*70}")
        
        # Create model (dropout supported)
        model = create_resnet18_model(num_classes=4, pretrained=True, dropout=exp['dropout'])
        
        # Setup optimizer (LR/OPT kept; add weight_decay)
        if exp['optimizer'] == 'Adam':
            optimizer = optim.Adam(model.parameters(), lr=exp['lr'], weight_decay=exp['weight_decay'])
        else:  # SGD
            optimizer = optim.SGD(model.parameters(), lr=exp['lr'], momentum=0.9, weight_decay=exp['weight_decay'])
        
        criterion = nn.CrossEntropyLoss()

        # No scheduler
        trained_model, history = train_model(
            model, train_loader, val_loader, criterion, optimizer, num_epochs, device,
            scheduler=None, scheduler_type='none'
        )
        
        # Store results
        best_val_acc = max(history['val_acc'])
        results.append({
            'name': exp['name'],
            'lr': exp['lr'],
            'optimizer': exp['optimizer'],
            'weight_decay': exp['weight_decay'],
            'dropout': exp['dropout'],
            'scheduler': 'none',
            'best_val_acc': best_val_acc,
            'final_val_acc': history['val_acc'][-1],
            'model': trained_model
        })
        all_histories[exp['name']] = history
        
        print(f"\n✓ Completed! Best Val Acc = {best_val_acc:.4f}")
    
    return results, all_histories

# ==================== EVALUATION FUNCTIONS ====================
def evaluate_model(model, dataloader, device):
    """Evaluate model and return predictions"""
    model.eval()
    all_preds, all_labels, all_probs = [], [], []
    with torch.no_grad():
        for images, labels in tqdm(dataloader, desc="Evaluating"):
            images = images.to(device)
            outputs = model(images)
            probs = torch.softmax(outputs, dim=1)
            _, predicted = torch.max(outputs, 1)
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.numpy())
            all_probs.extend(probs.cpu().numpy())
    return np.array(all_preds), np.array(all_labels), np.array(all_probs)

# ==================== VISUALIZATION FUNCTIONS ====================
def plot_training_history(history, title="Training History"):
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    axes[0].plot(history['train_loss'], label='Train Loss', marker='o', linewidth=2)
    axes[0].plot(history['val_loss'], label='Val Loss', marker='s', linewidth=2)
    axes[0].set_xlabel('Epoch', fontsize=12)
    axes[0].set_ylabel('Loss', fontsize=12)
    axes[0].set_title(f'{title} - Loss', fontsize=14, fontweight='bold')
    axes[0].legend(fontsize=10)
    axes[0].grid(True, alpha=0.3)
    axes[1].plot(history['train_acc'], label='Train Acc', marker='o', linewidth=2)
    axes[1].plot(history['val_acc'], label='Val Acc', marker='s', linewidth=2)
    axes[1].set_xlabel('Epoch', fontsize=12)
    axes[1].set_ylabel('Accuracy', fontsize=12)
    axes[1].set_title(f'{title} - Accuracy', fontsize=14, fontweight='bold')
    axes[1].legend(fontsize=10)
    axes[1].grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

def plot_hyperparameter_comparison(results):
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    names = [r['name'] for r in results]
    best_accs = [r['best_val_acc'] for r in results]
    final_accs = [r['final_val_acc'] for r in results]
    x = np.arange(len(names)); width = 0.35
    axes[0].bar(x - width/2, best_accs, width, label='Best Val Acc', alpha=0.8, color='steelblue')
    axes[0].bar(x + width/2, final_accs, width, label='Final Val Acc', alpha=0.8, color='coral')
    axes[0].set_xlabel('Configuration', fontsize=12)
    axes[0].set_ylabel('Accuracy', fontsize=12)
    axes[0].set_title('Hyperparameter Comparison', fontsize=14, fontweight='bold')
    axes[0].set_xticks(x); axes[0].set_xticklabels(names, rotation=45, ha='right')
    axes[0].legend(fontsize=10); axes[0].grid(True, alpha=0.3, axis='y')
    sorted_results = sorted(results, key=lambda x: x['best_val_acc'], reverse=True)
    sorted_names = [r['name'] for r in sorted_results]
    sorted_accs = [r['best_val_acc'] for r in sorted_results]
    colors = plt.cm.RdYlGn(np.linspace(0.3, 0.9, len(sorted_names)))
    axes[1].barh(sorted_names, sorted_accs, alpha=0.8, color=colors)
    axes[1].set_xlabel('Best Validation Accuracy', fontsize=12)
    axes[1].set_title('Ranked by Performance', fontsize=14, fontweight='bold')
    axes[1].grid(True, alpha=0.3, axis='x')
    for i, (name, acc) in enumerate(zip(sorted_names, sorted_accs)):
        axes[1].text(acc + 0.005, i, f'{acc:.4f}', va='center', fontsize=9)
    plt.tight_layout()
    plt.show()

def plot_all_training_curves(all_histories):
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    for name, history in all_histories.items():
        axes[0].plot(history['val_loss'], label=name, marker='o', markersize=4, linewidth=2)
        axes[1].plot(history['val_acc'], label=name, marker='o', markersize=4, linewidth=2)
    axes[0].set_xlabel('Epoch', fontsize=12); axes[0].set_ylabel('Validation Loss', fontsize=12)
    axes[0].set_title('Validation Loss - All Experiments', fontsize=14, fontweight='bold')
    axes[0].legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9); axes[0].grid(True, alpha=0.3)
    axes[1].set_xlabel('Epoch', fontsize=12); axes[1].set_ylabel('Validation Accuracy', fontsize=12)
    axes[1].set_title('Validation Accuracy - All Experiments', fontsize=14, fontweight='bold')
    axes[1].legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9); axes[1].grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(y_true, y_pred, class_names):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar_kws={'label': 'Count'},
                xticklabels=class_names, yticklabels=class_names, linewidths=0.5)
    plt.title('Confusion Matrix', fontsize=16, fontweight='bold', pad=20)
    plt.ylabel('True Label', fontsize=12); plt.xlabel('Predicted Label', fontsize=12)
    plt.xticks(rotation=45, ha='right'); plt.yticks(rotation=0)
    plt.tight_layout(); plt.show()
    return cm

def plot_sample_predictions(model, dataloader, device, class_names, num_samples=8):
    model.eval()
    images, labels = next(iter(dataloader))
    images_display = images[:num_samples]
    labels_display = labels[:num_samples]
    with torch.no_grad():
        outputs = model(images_display.to(device))
        probs = torch.softmax(outputs, dim=1)
        confidences, predictions = torch.max(probs, 1)
    fig, axes = plt.subplots(2, 4, figsize=(18, 9)); axes = axes.ravel()
    for i in range(num_samples):
        img = images_display[i].cpu().numpy().transpose(1, 2, 0)
        mean = np.array([0.485, 0.456, 0.406]); std = np.array([0.229, 0.224, 0.225])
        img = np.clip(std * img + mean, 0, 1)
        true_label = class_names[labels_display[i]]
        pred_label = class_names[predictions[i]]
        confidence = confidences[i].item()
        is_correct = predictions[i] == labels_display[i]
        color = 'green' if is_correct else 'red'
        border_color = 'lightgreen' if is_correct else 'lightcoral'
        axes[i].imshow(img); axes[i].axis('off')
        title = f'True: {true_label}\nPred: {pred_label}\nConf: {confidence:.2%}'
        axes[i].set_title(title, color=color, fontsize=11, fontweight='bold', pad=10)
        for spine in axes[i].spines.values():
            spine.set_edgecolor(border_color); spine.set_linewidth(3); spine.set_visible(True)
    plt.suptitle('Sample Predictions (Green=Correct, Red=Incorrect)', fontsize=14, fontweight='bold', y=0.98)
    plt.tight_layout(); plt.show()

# ==================== 6. EXECUTE TRAINING ====================
print("\n[STEP 6/6] Training Models with Hyperparameter Search (no scheduler)...")

# Run hyperparameter search (NO scheduler; LR/OPT kept the same)
results, all_histories = hyperparameter_search(
    trainLoader, valLoader, device, num_epochs=10
)

# Find best model
best_result = max(results, key=lambda x: x['best_val_acc'])
best_model = best_result['model']

print(f"\n{'='*70}")
print(f"🏆 BEST MODEL: {best_result['name']}")
print(f"   Best Validation Accuracy: {best_result['best_val_acc']:.4f}")
print(f"   Learning Rate: {best_result['lr']}")
print(f"   Optimizer: {best_result['optimizer']}")
print(f"   Weight Decay: {best_result['weight_decay']}")
print(f"   Dropout: {best_result['dropout']}")
print(f"   Scheduler: {best_result['scheduler']}")
print(f"{'='*70}\n")

# ==================== VISUALIZATIONS ====================
print("\nGenerating visualizations...")
plot_hyperparameter_comparison(results)
plot_training_history(all_histories[best_result['name']], title=f"Best Model: {best_result['name']}")
plot_all_training_curves(all_histories)

# ==================== FINAL EVALUATION ====================
print("\n" + "="*70)
print("FINAL EVALUATION ON TEST SET")
print("="*70 + "\n")

y_pred, y_true, y_probs = evaluate_model(best_model, testingLoader, device)
class_names = ["Mild_Demented", "Moderate_Demented", "Non_Demented", "Very_Mild_Demented"]

test_acc = accuracy_score(y_true, y_pred)
print(f"📊 Test Accuracy: {test_acc:.4f} ({test_acc*100:.2f}%)\n")

print("="*70)
print("CLASSIFICATION REPORT")
print("="*70)
print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# Confusion matrix & sample preds
cm = plot_confusion_matrix(y_true, y_pred, class_names)
plot_sample_predictions(best_model, testingLoader, device, class_names, num_samples=8)

# =============== F2 SCORES (β = 2) ===============
print("\n" + "="*70)
print("F2 SCORES (β = 2)")
print("="*70)
f2_macro = fbeta_score(y_true, y_pred, beta=2, average='macro')       # all classes equally
f2_weighted = fbeta_score(y_true, y_pred, beta=2, average='weighted') # weighted by support
f2_per_class = fbeta_score(y_true, y_pred, beta=2, average=None)      # per-class

print(f"F2 (macro):    {f2_macro:.4f}")
print(f"F2 (weighted): {f2_weighted:.4f}")
print("F2 (per class):")
for i, cname in enumerate(class_names):
    print(f"  {cname:20s}: {f2_per_class[i]:.4f}")

# Optional: bar chart for per-class F2
plt.figure(figsize=(8,4))
plt.bar(class_names, f2_per_class, alpha=0.85)
plt.title("Per-class F2 (β=2)")
plt.ylabel("F2 score")
plt.xticks(rotation=30, ha='right')
plt.ylim(0, 1.0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Per-class accuracy & summary
print("\n" + "="*70)
print("PER-CLASS ACCURACY")
print("="*70)
for i, class_name in enumerate(class_names):
    class_mask = y_true == i
    if class_mask.sum() > 0:
        class_acc = (y_pred[class_mask] == y_true[class_mask]).mean()
        class_count = class_mask.sum()
        print(f"{class_name:25s}: {class_acc:.4f} ({class_acc*100:.2f}%) - {class_count} samples")

print("\n" + "="*70)
print("SUMMARY STATISTICS")
print("="*70)
print(f"Total Test Samples: {len(y_true)}")
print(f"Correct Predictions: {(y_pred == y_true).sum()}")
print(f"Incorrect Predictions: {(y_pred != y_true).sum()}")
print(f"Overall Accuracy: {test_acc:.4f} ({test_acc*100:.2f}%)")

print("\n" + "="*70)
print("✅ TRAINING AND EVALUATION COMPLETE!")
print("="*70)

print("\n💾 Saving best model...")
torch.save(best_model.state_dict(), 'best_alzheimer_modelres.pth')
print("Model saved as 'best_alzheimer_modelres.pth'")



print("\n💾 Saving best model...")
torch.save(best_model.state_dict(), 'best_alzheimer_modelres.pth')
print("Model saved as 'best_alzheimer_modelres.pth'")

# --- ADD THESE LINES TO ALSO SAVE A NUMPY VERSION OF THE WEIGHTS ---
import numpy as np

# 1) Save the trained weights as a NumPy .npz archive
np_state = {k: v.detach().cpu().numpy() for k, v in best_model.state_dict().items()}
np.savez_compressed('best_alzheimer_modelres_np.npz', **np_state)
print("Also saved NumPy weights as 'best_alzheimer_modelres_np.npz'")

# 2) (Optional) Save the evaluation arrays too
np.save('y_true.npy',  y_true)
np.save('y_pred.npy',  y_pred)
np.save('y_probs.npy', y_probs)
print("Saved evaluation arrays: y_true.npy, y_pred.npy, y_probs.npy")
# --- END ADDITIONS ---
```
